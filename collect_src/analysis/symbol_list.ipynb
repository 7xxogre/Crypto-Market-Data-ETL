```python
import requests

res = requests.get("https://api.upbit.com/v1/market/all").json()
res
```

```python
all_symbol_list = []
for elem in res:
  if elem['market'].startswith("KRW"):
    sym = elem['market'][4:]
    all_symbol_list.append(sym)
all_symbol_list
```

```python
for idx, sym in enumerate(all_symbol_list):
    if idx % 10 == 0:
        print()

    print(f'"{sym}", ', end = '')
```
